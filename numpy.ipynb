```python
arr1=np.arange(10)
arr2=np.random.randn(10)
print("array 1\n",arr1,arr1.dtype)
print("array 2\n",arr2,arr2.dtype)
new=arr1.astype(arr2.dtype)
print("new array",new,new.dtype)
```

```python
arr=np.arange(12)
new=arr.reshape(4,3)
print(new)
new=arr.reshape((4,3),order='f')
print("\n",new)
```

```python
arr=np.array([[1,2,3,4,9,10],[5,6,7,8,11,12]])
print(arr)
new=arr.reshape(3,2,2)
print("\n",new)
new1=new.reshape(3,-1)
print("\n reshape(3,-1) \n",new1)
```

```python
arr=np.array([[10,20],[30,40]])
print("original:\n",arr)
flat=arr.flatten()
flat[0]=990
print("flattened:\n", flat)
print("original:\n",arr)
rav=arr.ravel()
rav[1]=888
print("raveled:\n",rav)
print("original:\n",arr)
```

```python
arr=np.array([[1,2],[3,4]])
arr2=np.array([[2,1],[3,5]])
gre=arr>arr2
less=arr<arr2
equal=arr==arr2
print(gre)
print(less)
print(equal)
```

```python
x=np.arange(10)
print("x= ",x)
print("\n x[:5]= ",x[:5])
print("\n x[5:]= ",x[5:])
print("\n x[::2]= ",x[::2])
print("\n x[5::-2]= ",x[5::-2])
```

```python
x=np.arange(10)
print(x)
x[5:]=2
print("\n after broadcasting1",x)

x[::2]=100
print("\n after broadcasting2",x)
```

```python
x=np.array([[12,5,2,4],[7,6,8,8],[1,6,7,7]])
print(x[0])
print(x[2,0])
print(x[1][2])
print(x[2,-1])
print(x[0])
```

```python
a=np.arange(12).reshape(3,4)
print("a= \n",a)
b=a>4
print("b= \n",b)
print("a[b]= \n",a[b])
a[b]=1
print("a= \n",a)
```

```python
arr=np.arange(12).reshape((3,4))
print(arr)
print("transpone .T")
print(arr.T)
print(np.transpose(arr))
```

```python
arr=np.arange(12).reshape((3,4))
print(np.dot(arr, arr.T))
```

```python
arr=np.arange(12).reshape((3,4))
print(arr)
print("swap 0 and 1:")
print(np.swapaxes(arr,0,1))
```

```python
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
result=np.multiply(arr1,arr2)
print(result)
result=np.matmul(arr1,arr2)
print(result)
```
