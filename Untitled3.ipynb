```python
# series indexing

obj = pd.Series(np.arange(4.), index=['a', 'b', 'c', 'd'])

print("obj['b']= \n", obj['b'])

print("obj[1]= \n", obj[1])

print("obj[2:4]= \n", obj[2:4])

print("obj[['b', 'a', 'd']]= \n", obj[['b', 'a', 'd']])

print("obj['b':'d']= \n", obj['b':'d'])  # no n-1 concept

print("obj[[1, 3]]= \n", obj[[1, 3]])

print("obj[obj < 2]= \n", obj[obj < 2])
```
