```python
# ===============================================================
# EXPERIMENT 4 â€“ Implementation of Perceptron Learning Algorithm
# ===============================================================

import numpy as np

# Input and desired output for AND logic gate
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

Y = np.array([0, 0, 0, 1])

# Initialize weights and parameters
w = np.array([0.3, -0.2])   # initial weights
theta = 0.4                 # threshold (bias)
lr = 0.2                    # learning rate
epochs = 5                  # number of epochs

print(f"Initial weights: {w}, theta = {theta}, lr = {lr}\n")

# Training process
for epoch in range(1, epochs + 1):
    print(f"--- Epoch {epoch} ---")
    print(f"x1\tx2\ty_des\ty_est\terror\tw1\tw2")
    error_count = 0

    for i in range(len(X)):
        x = X[i]
        y_des = Y[i]

        # Net input calculation
        net = np.dot(x, w) + theta
        y_est = 1 if net >= 0 else 0
        error = y_des - y_est

        # Weight update rule
        if error != 0:
            w = w + lr * error * x
            error_count += 1

        print(f"{x[0]}\t{x[1]}\t{y_des}\t{y_est}\t{error}\t{w[0]:.1f}\t{w[1]:.1f}")

    if error_count == 0:
        print("\nTraining complete!")
        break

# Display final perceptron outputs
print("\nFinal perceptron outputs:")
for i in range(len(X)):
    x = X[i]
    net = np.dot(x, w) + theta
    y_est = 1 if net >= 0 else 0
    print(f"Input {x} -> Net = {round(net, 1)}, Y_est = {y_est}, Y_des = {Y[i]}")
```
